```python
# https://ru.wikipedia.org/wiki/Гипотеза_Коллатца

import matplotlib.pyplot as plt

from time import time
```

```python
X_MIN = 1

X_MAX = 1000

MAX_ITERATIONS = 200

FIGSIZE = (17,7)

x_range = [i for i in range(X_MIN, X_MAX+1)]
```

```python
def modify_x(x_prev):
    X_next = 3 * x_prev + 1 if x_prev % 2 != 0 else x_prev//2
    return X_next


def iterate_x(X, max_iters):   
    x_values = [X]
    iters = [i for i in range(max_iters+1)]

    for itr in iters[1:]:
        X = modify_x(X)
        x_values.append(X)
    
    return iters, x_values
```

```python
sequences = []

# timer
start = time()

for x in x_range:
    x_iterated = iterate_x(x, MAX_ITERATIONS)
    sequences.append(x_iterated)

total_time = time() - start

print(f"Sequences:\t{len(sequences)}")
print(f"Total seconds:\t{round(total_time, 3)}")
```

```python
fig = plt.figure(figsize=FIGSIZE)    

for seq in sequences:
    plt.plot(seq[0], seq[1])
    
plt.show()
```
